```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
underlying_price_at_t0 = 100.
strike_price = 105.
time_to_maturity = 1.0
risk_free_rate = 0.05
underlying_volatility = 0.10
number_of_simulations = 500
```

```python
np.random.seed(1000)
z = np.random.standard_normal(number_of_simulations)
```

```python
simulation_stock_price_at_maturity = underlying_price_at_t0 * \
                                     np.exp((risk_free_rate-underlying_volatility**2/2)*time_to_maturity + \
                                     underlying_volatility * math.sqrt(time_to_maturity)* z) 
payoff_at_maturity = np.maximum(simulation_stock_price_at_maturity - strike_price, 0)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15, 6))
axes.plot(simulation_stock_price_at_maturity, linewidth = 1)
axes.plot(payoff_at_maturity, linewidth = 1)
```

```python
monte_carlo_estimator = math.exp(-risk_free_rate * time_to_maturity) * np.mean(payoff_at_maturity)
print("Value of the European call option: {:5.6f}.".format(monte_carlo_estimator))
```
